```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from chunk_utils import get_xpos_from_delays
```

```python
data_dir = './/data//'
matplotlib.rcParams['figure.figsize'] = 24,15
```

```python
df = pd.read_csv(data_dir+'cmu_dataset.csv')
df.head()
```

```python
u_indx = 1233
delays = np.array(df.iloc[:u_indx+32,3:])
username = df.iloc[u_indx,0]
typedname = '.tie5Roanl'
```

```python
username
```

```python
ypos = 0.01
ydelta = 0.0317

for dd in delays:
    xposvector = get_xpos_from_delays(dd, space_min = 0.006, max_delta_spacing=0.08)
    for char, xpos in zip(typedname, xposvector):
        plt.text(xpos, ypos + ydelta, char, fontsize=22)
    ypos += ydelta
plt.tight_layout()
plt.savefig(".//images//cmu//spatial_plot_" + username + ".svg", format='svg')
plt.close()
```
