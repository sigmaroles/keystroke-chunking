```python
import os
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
data_dir = './/data'
image_dir = './/images'
```

```python
# basic init
filenames = list(filter(lambda x: x.endswith('.kdt'), os.listdir(data_dir)))
matplotlib.rcParams['figure.figsize'] = 13,9
```

```python
from chunk_utils import kdata2delays
```

```python
# main program
for fn in filenames:
    with open(data_dir+'//'+fn, 'r') as fh:
        filedata = fh.read()
    typedname, rawkd = filedata.split('\n')
    n_chars = len(typedname)

    delays = kdata2delays(typedname, rawkd)
    
    # got the delays, now plot them
    
    xaxis = list(range(n_chars-1))
    for indx, de1 in enumerate(delays[-10:]):
        plt.plot(xaxis, de1, lw=0, markersize=5, marker='*', label="iteration "+str(indx+11))

    plt.legend()
    plt.xticks(np.arange(n_chars), [x for x in typedname])
    plt.title(typedname)
    plt.xlabel("Key", fontsize=15)
    plt.ylabel("Delay (flight time after)", fontsize=15)
    plt.tight_layout()
    
    plt.savefig(image_dir+'//gen_'+typedname+'.svg', format='svg', dpi='figure')
    plt.close()
    plt.clf()
    print ("Done with "+typedname)
```
